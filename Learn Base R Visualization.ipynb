# The Basics

```r
# Consider the whiteside dataset,
# which records the weekly gas consumption and average external temperature,
# as well as whether the measurements were before or after installing cavity-wall insulation.
# at Mr. Derek Whiteside's house in south-east England.
library(MASS)
library(dplyr)
glimpse(whiteside)
head(whiteside, 5)
tail(whiteside, 5)
```

```r
# Base R comes with the built-in plot() function,
# which is arguably the most straightforward way to create a data visualization.
# If we pass in a dataset with multiple variables, without specifying which columns we are looking at,
# We will get an array showing all possible relationships
plot(whiteside)

# For the whiteside dataset, this is fine, but for a dataset with many more columns,
# we may get an array with so much detail, it is overwhelming and difficult to interpret.
plot(mtcars)
```

```r
# Rather than plotting every possible relationship, we can select specific columns

# Plot gas consumption based on temperature.
plot(whiteside$Temp, whiteside$Gas)

# Plot gas consumption based on whether cavity-wall insulation was installed.
plot(whiteside$Insul, whiteside$Gas)
```

# Parameters

Base R offers several parameters to customize your visualization to exactly how you want it to look.

```r
# Start by plotting the max price and min price of cars in the Cars93 dataset from the MASS library.
library(MASS)
plot(Cars93$Max.Price, Cars93$Min.Price)
```

```r
# With the pch= parameter, we can specify how each point on the graph should look
par(mfrow=c(2,2))
plot(Cars93$Max.Price, Cars93$Min.Price, pch=1, main="pch=1")
plot(Cars93$Max.Price, Cars93$Min.Price, pch=16, main="pch=16")
plot(Cars93$Max.Price, Cars93$Min.Price, pch=2, main="pch=2")
plot(Cars93$Max.Price, Cars93$Min.Price, pch=25, main="pch=25")
```

```r
# With the col= parameter, we can speficy the color of each point.
par(mfrow=c(2,2))
plot(Cars93$Max.Price, Cars93$Min.Price, pch=1, main="pch=1", col="red") # specify color by name
plot(Cars93$Max.Price, Cars93$Min.Price, pch=16, main="pch=16", col="red") # specify color by name
plot(Cars93$Max.Price, Cars93$Min.Price, pch=2, main="pch=2", col="4") # specify color by number
plot(Cars93$Max.Price, Cars93$Min.Price, pch=25, main="pch=25", col="#ace1af") # specify color by hexideciamal
```

# Supplementary Functions for Visualiations

```r
# By using the points() function, we can add a new set of points to the graph.
# Here we will examine the price of a car compared to the luggage room.
par(mfrow=c(1,2))
plot(Cars93$Luggage.room, Cars93$Price, pch=1, col="red")
# In this graph, a clear relationship is not obvious.

# Now, on the same plot, we we add points showing the relationship between fuel tank capacity and car length.
plot(Cars93$Luggage.room, Cars93$Price, pch=1, col="red")
points(Cars93$Fuel.tank.capacity, Cars93$Price, pch=2, col="blue")
# The relationship between fuel tank capacity and price is visibly clearer,
# han that between price and luggage space.

# cor() function to confirm
cor(Cars93$Luggage.room, Cars93$Price, use="complete.obs")
cor(Cars93$Fuel.tank.capacity, Cars93$Price)
```

```r
# With the abline() function, we can manually add a reference line,
# where a is the y-intercept, and b is the slope.
plot(Cars93$Luggage.room, Cars93$Price, pch=1, col="red")
points(Cars93$Fuel.tank.capacity, Cars93$Price, pch=2, col="blue")
abline(a=-20, b=2.5)
# With this reference line,
# it is much easier a linear relationship between fuel tank capacity & price, than between luggage room & price.
```

```r
# Instead of definine the slope and y-intercept on our own,
# we can use the lm() function to create a linear model for us.
plot(Cars93$Luggage.room, Cars93$Price, pch=1, col="red")
points(Cars93$Fuel.tank.capacity, Cars93$Price, pch=2, col="blue")

# We can also use the lty= parameter to define the line type.
# Here, we will examine how well the reference line actually follows the data in each relationship.
abline(lm(Price ~ Fuel.tank.capacity, data=Cars93), lty=1, col="blue")
abline(lm(Price ~ Luggage.room, data=Cars93), lty=2, col="red")
```
