# The Basics

```r
# Consider the whiteside dataset,
# which records the weekly gas consumption and average external temperature,
# as well as whether the measurements were before or after installing cavity-wall insulation.
# at Mr. Derek Whiteside's house in south-east England.
library(MASS)
library(dplyr)
glimpse(whiteside)
head(whiteside, 5)
tail(whiteside, 5)
```

```r
# Base R comes with the built-in plot() function,
# which is arguably the most straightforward way to create a data visualization.
# If we pass in a dataset with multiple variables, without specifying which columns we are looking at,
# We will get an array showing all possible relationships
plot(whiteside)

# For the whiteside dataset, this is fine, but for a dataset with many more columns,
# we may get an array with so much detail, it is overwhelming and difficult to interpret.
plot(mtcars)
```

```r
# Rather than plotting every possible relationship, we can select specific columns

# Plot gas consumption based on temperature.
plot(whiteside$Temp, whiteside$Gas)

# Plot gas consumption based on whether cavity-wall insulation was installed.
plot(whiteside$Insul, whiteside$Gas)
```
